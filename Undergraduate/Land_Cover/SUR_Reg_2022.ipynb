```python
# Extract TAEP
def extract_data_final(sheet_name, file_path):
    # Read the Excel sheet without setting any header, then manually set the header
    df_TAEP = pd.read_excel(file_path, sheet_name=sheet_name, header=None)
    
    # Drop unnecessary rows and reset the index
    df_TAEP = df_TAEP.drop([0, 1, 2]).reset_index(drop=True)
    
    # Extract relevant data and format
    df_extracted = df_TAEP.iloc[:, [0, 4]].copy()
    df_extracted.columns = ['County', 'Variable']
    df_extracted['County'] = df_extracted['County'].str.title() + " County"
    df_extracted['Year'] = int(sheet_name.split('-')[1]) + 2000  # Extract year from sheet name
    
    return df_extracted

# Extract data again for each year from 2010 to 2021
df_TAEP_list = [extract_data_final(f"TAEP-{str(year)[2:]}", file_path) for year in range(2010, 2023)] # 2022 data!!

# Concatenate all extracted data
df_TAEP = pd.concat(df_TAEP_list, axis=0, ignore_index=True)

df_TAEP.head()
```

```python
# check if household count, population, or both is best:

def run_model_with_variables(data, variables):
    equations = {
        'Total_Crops': f'Total_Crops ~ 1 + Pasture + Hay + Forest + Developed_Open + Developed_Non_Open + Log_Median_Income + {variables} + Investment_1000s + {year_cols}',
        'Pasture': f'Pasture ~ 1 + Total_Crops + Hay + Forest + Developed_Open + Developed_Non_Open + Log_Median_Income + {variables} + Investment_1000s + {year_cols}',
        'Hay': f'Hay ~ 1 + Pasture + Total_Crops + Forest + Developed_Open + Developed_Non_Open + Log_Median_Income + {variables} + Investment_1000s + {year_cols}',
        'Forest': f'Forest ~ 1 + Pasture + Hay + Total_Crops + Developed_Open + Developed_Non_Open + Log_Median_Income + {variables} + Investment_1000s + {year_cols}',
        'Developed_Open': f'Developed_Open ~ 1 + Pasture + Hay + Forest + Total_Crops + Developed_Non_Open + Log_Median_Income + {variables} + Investment_1000s + {year_cols}',
        'Developed_Non_Open': f'Developed_Non_Open ~ 1 + Pasture + Hay + Forest + Developed_Open + Total_Crops + Log_Median_Income + {variables} + Investment_1000s + {year_cols}',
    }

    mod = SUR.from_formula(equations, data=data)
    res = mod.fit(cov_type='unadjusted')
    return res

# Model with Household_Count_1000s
res_household = run_model_with_variables(merged_df, 'Household_Count_1000s')
print("Model with Household_Count_1000s:")
print(res_household)

# Model with Population_1000s
res_population = run_model_with_variables(merged_df, 'Population_1000s')
print("Model with Population_1000s:")
print(res_population)

# Model with both Household_Count_1000s and Population_1000s
res_both = run_model_with_variables(merged_df, 'Household_Count_1000s + Population_1000s')
print("Model with both Household_Count_1000s and Population_1000s:")
print(res_both)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from linearmodels.system import SUR

def calculate_vif(df, variables):
    vif_data = pd.DataFrame()
    vif_data["feature"] = variables
    vif_data["VIF"] = [variance_inflation_factor(df[variables].values, i) for i in range(len(variables))]
    return vif_data

def create_lagged_variables(data):
    for lag in range(1, 4):  # Create lags up to 3
        data[f'Investment_1000s_lag{lag}'] = data.groupby('County')['Investment_1000s'].shift(lag)

def run_model_with_lags(data, lags, year_cols):
    # Include original investment variable only if 0 is in lags
    investment_var = 'Investment_1000s' if 0 in lags else ''

    # Create lag terms, excluding 0
    lag_terms = ' + '.join([f'Investment_1000s_lag{lag}' for lag in lags if lag != 0])

    equations = {
    'Total_Crops': f'Total_Crops ~ 1 + Pasture + Hay + Forest + Developed_Open + Developed_Non_Open + Log_Median_Income + Household_Count_1000s + Population_1000s + {investment_var} + {lag_terms} + {year_cols}',
    'Pasture': f'Pasture ~ 1 + Total_Crops + Hay + Forest + Developed_Open + Developed_Non_Open + Log_Median_Income + Household_Count_1000s + Population_1000s + {investment_var} + {lag_terms} + {year_cols}',
    'Hay': f'Hay ~ 1 + Pasture + Total_Crops + Forest + Developed_Open + Developed_Non_Open + Log_Median_Income + Household_Count_1000s + Population_1000s + {investment_var} + {lag_terms} + {year_cols}',
    'Forest': f'Forest ~ 1 + Pasture + Hay + Total_Crops + Developed_Open + Developed_Non_Open + Log_Median_Income + Household_Count_1000s + Population_1000s + {investment_var} + {lag_terms} + {year_cols}',
    'Developed_Open': f'Developed_Open ~ 1 + Pasture + Hay + Forest + Total_Crops + Developed_Non_Open + Log_Median_Income + Household_Count_1000s + Population_1000s + {investment_var} + {lag_terms} + {year_cols}',
    'Developed_Non_Open': f'Developed_Non_Open ~ 1 + Pasture + Hay + Forest + Developed_Open + Total_Crops + Log_Median_Income + Household_Count_1000s + Population_1000s + {investment_var} + {lag_terms} + {year_cols}',
}

    data_clean = data.dropna()
    
    mod = SUR.from_formula(equations, data=data_clean)
    res = mod.fit(cov_type='robust')
    r_squared = res.rsquared

    # Prepare variables for VIF calculation
    all_vars = 'Pasture + Hay + Forest + Developed_Open + Developed_Non_Open + Log_Median_Income + Household_Count_1000s + Population_1000s + ' + investment_var + ' + ' + lag_terms + ' + ' + year_cols
    vif_variables = all_vars.replace('+', '').split()

    # Calculate VIF
    vif_df = calculate_vif(data[vif_variables], vif_variables)

    return r_squared, data.shape, vif_df

create_lagged_variables(merged_df)  # Create lagged variables

# Define combinations
lag_combinations = {
    't0': [0],
    't0_t1': [0, 1],
    't0_t1_t2': [0, 1, 2],
    't0_t1_t2_t3': [0, 1, 2, 3]
}

# Run models and store results
for combo_name, lags in lag_combinations.items():
    r_squared, df_shape, vif_df = run_model_with_lags(merged_df, lags, year_cols)
    print(f"{combo_name}: R-squared = {r_squared}, DataFrame Shape: {df_shape}")
    print("VIF Results:")
    print(vif_df)
    print("\n")

# colinearity problems are seemingly guarnanteed with distributed lag
```
